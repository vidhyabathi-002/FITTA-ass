class method 




```python
class training:
    iname = 'FITA'
    head = 'ram'
    location = 'plk'

    # constructor
    def __init__(self, name, age, deg, yop, mark):
        self.name = name
        self.age = age
        self.deg = deg
        self.yop = yop
        self.mark = mark

    # object method
    def studentdetails(self):
        print(self.name, self.age, self.deg, self.yop, self.mark)

    # class method
    @classmethod
    def instituedetails(cls):
        print(cls.iname, cls.head, cls.location)
    # static method

    @staticmethod
    def reversename(name):
        name = name[::-1]
        return name
        




# object creation
s1 = training('vidhya', 20, 'b.tech', 2026, 94)
s1.studentdetails()

s2 = training('sri', 21, 'b.tech', 2025, 90)
s2.studentdetails()

s3 = training('monya', 22, 'b.tech', 2024, 92)
s3.studentdetails()

# class method call
training.instituedetails()

# static method call
print(training.reversename('vidhya'))
```
