normal public classes


```python
class Bank:
    bank_name = "icici"
    location = "pal"

    def __init__(self, name, acc_no, balance, pin):
        self.name = name
        self.acc_no = acc_no
        self.balance = balance
        self.pin = pin

    def show_dt(self):
        print("\n user details \n")
        print(f"name is {self.name}")
        print(f"acc_no is :{self.acc_no}")
        print(f"balance is :{self.balance}")
        print(f"pin is :{self.pin}")

obj1 = Bank("vidhya", 1234, 10000, 222)
obj1.show_dt()
```

protected method


```python
class Bank:
    bank_name = "icici"
    location = "pal"

    def __init__(self, name, acc_no, balance, pin):
        self.name = name
        self._acc_no = acc_no
        self.balance = balance
        self._pin = pin

    def show_dt(self):
        print("\n user details \n")
        print(f"name is {self.name}")
        print(f"acc_no is :{self._acc_no}")
        print(f"balance is :{self.balance}")
        print(f"pin is :{self._pin}")

obj1 = Bank("vidhya", 1234, 10000, 222)
obj1.show_dt()

print(obj1._pin)
obj1._pin = 144
print(obj1._pin)
```

getters and setters

Getter methods read private data. Setter methods update private data with controlled access.


```python
class Student:
    def __init__(self):
        self.__mark = 0

    def set_mark(self, mark):
        self.__mark = mark

    def get_mark(self):
        return self.__mark

s = Student()
s.set_mark(95)
print(s.get_mark())
```

```python
class Employee:
    def __init__(self):
        self.__salary = 0

    def set_salary(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary

obj1 = Employee()
obj1.set_salary(300)
print(obj1.get_salary())
```
