inheritance  concepts :

       Inheritance is an Object-Oriented Programming (OOP) concept where a child class inherits properties and methods from a parent class. This allows code reuse and creates a hierarchy of classes.

types of inheritance :

    1. Single Inheritance
           >>>>> One child class inherits from one parent class.

    2. Multiple Inheritance:
           >>>>> One child class inherits from multiple parent classes.

    3. Multilevel Inheritance
           >>>>> A chain of inheritance - grandparent â†’ parent â†’ child.

    4. Hierarchical Inheritance
          >>>>> one parent class has multiple child classes.


    5. Hybrid Inheritance
          >>>>>>A combination of two or more types of inheritance.

1.creat a parent class name , person , display name and age  and display child class emp 
the innerite person  sal display all detailes 


```python
class  person :#  parent class
    name="rio"
    age= 12


# print statement inside the parent class
    def person_info(self):
        print("\n this is parent class detailes ")
        print(f"name of the person is :{self.name}")
        print(f"age of the person is :{self.age}")

# print statement inside the child class    
class employee(person):
    emp_id =12345
    emp_dept="mech"
    emp_salary= 50000


# print statement inside the child class
    def emp_info(self):
        print("\n this is child class detailes")
        print(f"employee id is :{self.emp_id}")
        print(f"employee dept is :{self.emp_dept}")
        print(f"employee sal info :{self.emp_salary}")



# creating object for the child class   
obj1=employee()

obj1.person_info()# call the parent class inside the child class
obj1.emp_info()# call the child class
```

```python
class paranthaman_father:
    gf_name="thevaneya pavanar"
    age=1008
    dress_code="langotta and turban"
    wife_count=5

    def grandfather_info(self):
        print("\n this is grandfather detailes")
        print(f"grandfather name :{self.gf_name}")
        print(f"grandfather age :{self.age}")
        print(f"grandfather dress code :{self.dress_code}")
        print(f"grandfather wife count :{self.wife_count   }")
# creating object for the child class
class paranthaman_own(paranthaman_father):
    name="paranthaman"
    age= 30
    dress_code="pant and shirt"
    wife_count=1

    #defining method for the child class
    def paranthaman_info(self):
        print("\n this is paranthaman detailes")
        print(f"paranthaman name :{self.name}")
        print(f"paranthaman age :{self.age}")
        print(f"paranthaman dress code :{self.dress_code}")
        print(f"paranthaman wife count :{self.wife_count}")


class paranthaman_son(paranthaman_own):
    name="santhoshkumar p"
    age=22
    dress_code="saree with blouse"
    gender="transgender (or) gay"

    def santhoskumar_info(self):
        print("\n this is santhoshkumar detailes")
        print(f"santhoshkumar name :{self.name}")
        print(f"santhoshkumar age :{self.age}")
        print(f"santhoshkumar dress code :{self.dress_code}")
        print(f"santhoshkumar gender :{self.gender  }") 



obj1=paranthaman_son()
obj1.grandfather_info()# call the grand father class inside the child class
obj1.paranthaman_info()# call the paranthaman class inside the child class      
obj1.santhoskumar_info()# call the santhoshkumar class inside the child class
```
