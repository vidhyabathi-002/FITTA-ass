```python
# 9x9 Sudoku Generator

base = 3
side = base * base

# Pattern for a valid Sudoku
def pattern(r, c):
    return (base * (r % base) + r // base + c) % side

# Numbers 1 to 9
nums = [1,2,3,4,5,6,7,8,9]

# Create Sudoku grid
sudoku = []

for r in range(side):

    row = []

    for c in range(side):

        row.append(nums[pattern(r, c)])

    sudoku.append(row)

# Print Sudoku
for r in range(side):

    if r % 3 == 0 and r != 0:
        print("-" * 21)

    for c in range(side):

        if c % 3 == 0 and c != 0:
            print("|", end=" ")

        print(sudoku[r][c], end=" ")

    print()
```

```python
# Simple Sudoku Solver

grid = [
[5,3,0,0,7,0,0,0,0],
[6,0,0,1,9,5,0,0,0],
[0,9,8,0,0,0,0,6,0],

[8,0,0,0,6,0,0,0,3],
[4,0,0,8,0,3,0,0,1],
[7,0,0,0,2,0,0,0,6],

[0,6,0,0,0,0,2,8,0],
[0,0,0,4,1,9,0,0,5],
[0,0,0,0,8,0,0,7,9]
]


# Print Grid
def print_grid():

    for row in grid:
        print(row)


# Check Number
def is_valid(num, row, col):

    # Row check
    for i in range(9):

        if grid[row][i] == num:
            return False

    # Column check
    for i in range(9):

        if grid[i][col] == num:
            return False

    return True


# Solve Sudoku
for row in range(9):

    for col in range(9):

        if grid[row][col] == 0:

            for num in range(1,10):

                if is_valid(num, row, col):

                    grid[row][col] = num

                    break


print("Solved Grid:\n")

print_grid()
```
