1. STUDENT DETAILES      :create a class student with constructor to intialize name aand age, display the details

```python
class student :
    def __init__(self,name,age):
        self.name= name 
        self.age=age 

    def show_std_dt(self):
        print("\n STUDENT DETAILES")
        print(f'student name is :{self.name}')
        print(f'student age is: {self. age }')




obj1=student("vidhya",99)
obj1.show_std_dt()
```

2. CAR INFORMATION        :create a car with constructor to initialize brand and price , print car detailes

```python
class car:
    def __init__(self,car_brand,price):
        self.car_brand=car_brand
        self.price=price 

    def show_dt(self):
        print(f'car brand is : {self.car_brand} and price is :{self.price }')

obj=car("TATA","RS.1000000000")
obj.show_dt()
```

3. BOOK DETAILES             : create a class book  with constrector  initialize title and  author display book info

```python
class book :
    def __init__(self,title,author):
        self.title= title 
        self.author= author 

    def book_dt(self):
        print("\n  BOOK DETAILES")
        print(f'title name is :{self.title}')
        print(f'author NAME  is: {self. author }')




obj1=book("LOVE HYPOTHESIS","NOLAN")
obj1.book_dt()
```
