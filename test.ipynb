```python
# vowel and conso and spl and digit
s = input("enter the str:")
v=0
c=0
d=0
spl=0

for i in s:
    if i.isalpha():
        if i in "AEIOUaeiou":
            v+=1
        else:
            c+=1

    elif "0">=i>="9":
        d+=1

    else:
        spl=spl+1

print(s)
print("vowels:",v)
print("consonent:",c)
print("digit:",d)
print("special char:",spl)


```
