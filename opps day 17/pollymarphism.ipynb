1. Method Overriding
Definition

Method Overriding means:

A child class changes the implementation of a method already present in the parent class.

The method name and parameters are usually the same.

```python
# Parent Class
class Animal:
    
    # Parent method
    def sound(self):
        print("Animals make sounds")


# Child Class - Dog
class Dog(Animal):
    
    # Overriding parent method
    def sound(self):
        print("Dog barks: Bow Bow")


# Child Class - Cat
class Cat(Animal):
    
    # Overriding parent method
    def sound(self):
        print("Cat meows: Meow Meow")
        
# Creating objects
ani = Animal()
dog = Dog()
cat = Cat()

# Calling methods
ani.sound()
dog.sound()
cat.sound()
```

Explanation
Animal class already has sound()
Dog class redefines the same sound()
Child class method replaces parent class method

This is called Method Overriding.


2. Method Overloading
Definition

Method Overloading means:

Same method name with different number or type of arguments.

Python does not support true method overloading directly like Java/C++.

We achieve it using:

Default arguments
*args
