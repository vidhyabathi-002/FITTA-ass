01.Shape ProgramWrite a Python program using abstraction for Shape.Create an abstract class ShapeCreate an abstract method area()Create a child class SquareHint: area * area

```python
from abc import ABC, abstractmethod


# Abstract Class
class Shape(ABC):

    # Abstract Method
    @abstractmethod
    def area(self):
        pass


# Child Class
class Square(Shape):

    # Constructor
    def __init__(self, side):
        self.side = side

    # Implementing abstract method
    def area(self):
        print("Area of Square =", self.side * self.side)


# Object Creation
s = Square(4)

# Method Call
s.area()
```
