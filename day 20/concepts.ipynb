###alternative even-odd genrator 


```python
def std():
    for i in range(1, 10):
        yield i

x = std()

for i in x:
    if i%2==0:
        print(i,"even")
    else:
        print(i,"odd")
```

```python
###prime number in infinite usung genrator 


def prime():

    num = 2

    while True:

        check = True

        for i in range(2, num):

            if num % i == 0:
                check = False
                break

        if check:
            yield num

        num += 1


p = prime()

print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
    
```

```python

###prime number in infinite 

num = 2

while num <= 30:

    prime = True

    for i in range(2, num):

        if num % i == 0:
            prime = False
            break

    if prime:
        print(num)

    num += 1
```
