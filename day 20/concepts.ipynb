```python
def genrator():
    yield 1
    yield 2
    yield 3

gen= genrator()

for i in gen:
    print ("\n",i,"\n")
```

```python
def std_roll():
    roll=1

    while roll <=5:
        yield roll
        roll +=1

for i in std_roll():
    print(i)

```

```python
def std():
    for i in range(1, 20):
        yield i

x = std()

for i in x:
    print(i)
```

###alternative even-odd genrator 


```python
def std():
    for i in range(1, 10):
        yield i

x = std()

for i in x:
    if i%2==0:
        print(i,"even")
    else:
        print(i,"odd")
```

```python
def  decorator_fun(org_fun):
    def wrapper():
        print("welcome")
        org_fun()
        print("tank you")

    return wrapper

def display():
    print("hello std")
decorator=decorator_fun(display)
decorator()
```

```python
#### @ decorator method :

def  decorator_fun(org_fun):
    def wrapper():
        print("welcome")
        org_fun()
        print("tank you")

    return wrapper
@decorator_fun
def display():
    print("hello std")
display()
```
